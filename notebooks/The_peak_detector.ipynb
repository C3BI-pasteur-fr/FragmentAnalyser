In this notebook, we use a single data set and explore the effects of the parameter sigma, lower_bound, upper_bound and guess.

The algorithm that searches for peak uses the following algorithm:

X contains the positions of the peaks that ranges from 1 to 6000
Y contains the proportion of such peaks.

We are interested in identifying the position of the peak with maximum value but also in a valid range. 

We multiply Y by a gaussian centered around mu with a sigma of 50 (default). The value mu is infered from the median of each max found in the N wells. The value mu may also be provided by the user (guess parameter)


```python
from fragment_analyser import Line, fa_data
%pylab  inline
```

```python
filename = fa_data("alternate/peaktable.csv")
```

```python
l = Line(filename)
```

```python
l.diagnostic()
```

```python
# In the original data, there is a peak around x=65, stronger than the
# real peak we are looking for around x=600
# a cutoff at x=120 is used but can be changed if needed:
l = Line(filename, lower_bound=1)
l.diagnostic()
```

```python
# data is multiply by a gaussian centered around a value that is 
# either provided by the user or guessed from the median of the 
# max peaks across all wells. 
l = Line(filename, sigma=50)
l.set_guess(500)
l.diagnostic()
```

```python
# a wrong guess but large sigma works
l = Line(filename, sigma=100)
l.set_guess(500)
l.diagnostic()
```

```python
# if the data is made of similar experiments, you can decrease the sigma
l = Line(filename, sigma=10)
l.set_guess(600)
l.diagnostic()
```

```python
# Here, we want to get the largest peaks around 1200 if there are any:
l = Line(filename)
l.set_guess(1200)
l.diagnostic()
```
